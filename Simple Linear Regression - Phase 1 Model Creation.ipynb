```python
# Import pandas library to read and handle dataset files such as CSV

import pandas as pd
```

```python
# Load the dataset from the CSV file into a dataframe

dataset=pd.read_csv("Salary_Data.csv")
```

```python
# Display the dataset to understand its structure
#YearsExperience → input feature
#Salary → output value we want to predict.

dataset
```

```python
# Extract the independent variable (input feature)
#YearsExperience = independent variable (X)

independent=dataset[["YearsExperience"]]
independent
```

```python
# Extract the dependent variable (target/output)
#Salary is the target variable (Y) that the model will learn to predict.

dependent=dataset[["Salary"]]
dependent
```

```python
# Split dataset into training and testing sets
# 70% of data used for training and 30% used for testing
#test_size = 0.30 → 30% data reserved for testing.
#random_state = 0 → ensures the same split every time the code runs.

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
# Import Linear Regression algorithm

from sklearn.linear_model import LinearRegression

# Create the regression model
#Salary=w(YearsExperience)+b

regressor=LinearRegression()

# Train the model using the training dataset

regressor.fit(X_train,Y_train)
```

```python
#Retrieve the slope (weight) learned by the model
#Every 1 year of experience increases salary by about 9360

weight=regressor.coef_
weight
```

```python
# Retrieve the intercept (bias)
#Intercept is the predicted salary when experience = 0.

bias=regressor.intercept_
bias
```

```python
# Predict salary values for the test dataset

Y_pred=regressor.predict(X_test)
```

```python
# Import evaluation metric

from sklearn.metrics import r2_score

# Calculate R² score to measure model accuracy

r_score=r2_score(Y_test,Y_pred)
```

```python
r_score
```

```python
# Import pickle library to save and load trained machine learning models

import pickle

# Define the filename where the trained model will be stored
#.sav is just a common extension used for saved models

filename="finalized_model_linear.sav"
```

```python
# Save the trained regression model into a file using pickle

pickle.dump(regressor,open(filename,'wb'))

#This step stores the trained model on disk.
#pickle.dump() saves an object
#regressor trained ML model
#open(filename,'wb') open file in write binary mode
```

```python
# Load the saved model from the file
#This step retrieves the saved model from the file so it can be used again.
#pickle.load() loads serialized object
#rb read binary
#loaded_model loaded_model

loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))

# Predict salary for a given years of experience (10 years)
#Salary=9360.26×Experience+26777.39

result=loaded_model.predict([[10]])
```
