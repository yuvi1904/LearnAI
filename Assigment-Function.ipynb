```python
class AIField:
    def Subfield(self):
        requirement=input("Enter your requirement: ").strip().lower()
        if requirement=='aitech':
            return """Techs in AI are:
            Mechanical Tech
            Medical Tech
            Electronic Tech
            Industrial Tech
            Manufacturing Tech"""
        elif requirement=='subfields':
            return """Sub-fields in AI are:
            Machine Learning
            Neural Networks
            Vision
            Robotics
            Speech Processing
            Natural Language Processing"""
        else:
            return "Error! Enter a valid input"
```

```python
obj=AIField()
print(obj.Subfield())
```

```python
def oddEven():
    num=(int(input("Enter a number: ")))
    if((num%2)==0):
        print(num, "is Even Number")
        value="is Even Number"
    else:
        print(num, "is Odd Number")
        value="is Odd Number"
    return value
```

```python
def ElegibilityForMarriage():
    gender=input("Enter your gender(Male/Female/Other): ")
    age=int(input("Enter your age: "))
    if gender in ("male"):
        min_age=21
    elif gender in ("female"):
        min_age=18
    else:
        min_age=20
    status="Eligible" if age >= min_age else "Not Eligible"
    return f"Your Gender: {gender}\nYour Age: {age}\n{status}"
```

```python
result=ElegibilityForMarriage()
print(result)
```

```python
class FindPercent:
    def percentage(self):
        marks=[]
        
        for subjects in range(5):
            mark=int(input(f"Subject {subjects+1}: "))
            marks.append(mark)
            
        Total=sum(marks)
        Percentage=(Total/500)*100
        return f"You entered: {marks}\nTotal: {Total}\nYour Percentage: {Percentage}"
```

```python
object=FindPercent()
print(object.percentage())
```

```python
class Triangle:
    def TriangleCal(self):
        Height=int(input("Enter the Height: "))
        Breadth=int(input("Enter the Breadth: "))
        Area=(Height*Breadth)/2
        print(f"Area of the Triangle is {Area}")

        Height1=int(input("Enter the Height1: "))
        Height2=int(input("Enter the Height2: "))
        Height3=int(input("Enter the Height3: "))
        Perimeter=Height1+Height2+Height3
        print(f"Area of the Triangle is {Perimeter}")

        return Area, Perimeter
```
