```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int, drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend','Administration','Marketing Spend','State_Florida','State_New York']]
```

```python
independent
```

```python
dependent=dataset[['Profit']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='poisson', splitter='best')
regressor=regressor.fit(X_train,Y_train)
```

```python
#plot_tree() is a function inside sklearn.tree module(.tree module contains DecisionTreeClassifier, DecisionTreeClassifier, plot_tree() & plot_tree() & export functions), not inside the model itself
#tree.plot_tree(regressor) takes the trained mode(above) and converts it into a visual tree
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
Prediction=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Prediction)
```
